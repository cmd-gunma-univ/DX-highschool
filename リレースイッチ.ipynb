```python
import pigpio  # pigpioライブラリの読み込み
import time    # 時間操作のためのライブラリ

# pigpioの初期化
pi = pigpio.pi()  # pigpioの準備

# GPIO 27を出力に設定
GPIO_PIN = 27
pi.set_mode(GPIO_PIN, pigpio.OUTPUT)

try:
    for i in range(10):  # 点滅を10回繰り返す
        pi.write(GPIO_PIN, 1)  # GPIO 27をON
        time.sleep(10)          # 10秒間待機
        pi.write(GPIO_PIN, 0)  # GPIO 27をOFF
        time.sleep(5)          # 5秒間待機
finally:
    pi.write(GPIO_PIN, 0)  # 終了時にGPIOをOFF
    pi.stop()              # pigpioの終了
```
